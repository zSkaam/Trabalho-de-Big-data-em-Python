```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

dados = {
    'empresas': ['Grupo I - Superior a R$ 50 Milhões', 
                 'Grupo II - R$ 20 Milhões a R$ 50 Milhões', 
                 'Grupo III - R$ 6 Milhões a R$ 20 Milhões', 
                 'Grupo IV - R$ 4,8 Milhões a R$ 6 Milhões',
                 'Empresa de Pequeno Porte', 
                 'Microempresa'], 
    'serviços_prestados': [127, 98, 59, 30, 21, 14]
}

df_contagem = pd.DataFrame(dados)

print("DataFrame de estatistica de serviços prestados:")
print(df_contagem)

nome_arquivo_csv = 'Porte_das_empresas.csv'
df_contagem.to_csv(nome_arquivo_csv, index=False) 

print(f"\nDados salvos com sucesso em '{nome_arquivo_csv}'")

plt.figure(figsize=(10, 5))

sns.barplot(
    x='serviços_prestados', 
    y='empresas',           
    data=df_contagem, 
    palette='viridis'
)

plt.title('Distribuição de Serviços Prestados por Porte de Empresa', fontsize=16)
plt.xlabel('Número de Serviços Prestados')
plt.ylabel('Porte da Empresa')

plt.tight_layout() 

nome_arquivo_pdf = 'Porte_empresas_servicos.pdf'
plt.savefig(nome_arquivo_pdf, dpi=300)
print(f"\nGráfico salvo em: '{nome_arquivo_pdf}'")

plt.show()
```
