```python
import pandas as pd
import matplotlib.pyplot as plt

dados = {
    'contatos': ['Whatsapp', 'Telefone', 'Instagram', 'Twitter(x)', 'Facebook'],
    'acessos': [1347, 234, 671, 320, 431]
}

df_contagem = pd.DataFrame(dados)

print("DataFrame de Contagem de Acessos:")
print(df_contagem)

nome_arquivo_csv = 'meiosdecontato.csv'
df_contagem.to_csv(nome_arquivo_csv, index=False) 

print(f"\nDados salvos com sucesso em '{nome_arquivo_csv}'")

fig, ax = plt.subplots(figsize=(8, 8)) 

ax.pie(
    df_contagem['acessos'], 
    labels=df_contagem['contatos'], 
    autopct='%1.1f%%', 
    startangle=90, 
    wedgeprops={'edgecolor': 'black'}
)

ax.set_title('Distribui√ß√£o de Acessos\npor Meio de Contato', fontsize=15)
ax.axis('equal')

nome_arquivo_pdf = 'contatos_grafico_de_pizza.pdf'
fig.savefig(nome_arquivo_pdf, bbox_inches='tight')

print(f"\nO gr√°fico de pizza foi salvo como '{nome_arquivo_pdf}' no formato PDF.")

plt.show()
```
