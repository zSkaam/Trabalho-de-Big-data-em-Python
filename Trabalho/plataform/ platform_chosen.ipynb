```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud 

dados = {
    'plataformas': ['Whatsapp', 'Telefone', 'Instagram', 'Twitter(x)','Sites','Facebook'],
    'plataformas escolhidas': [1347, 234, 671, 320, 521, 276]
}

df_contagem = pd.DataFrame(dados)

print("DataFrame de plataformas:")
print(df_contagem)

nome_arquivo_csv = 'platform.csv'
df_contagem.to_csv(nome_arquivo_csv, index=False) 

print(f"\nDados salvos com sucesso em '{nome_arquivo_csv}'")

word_freq = dict(zip(df_contagem['plataformas'], df_contagem['plataformas escolhidas']))

wc = WordCloud(
    width=800, 
    height=400, 
    background_color='white', 
    colormap='viridis', 
    max_words=100 
).generate_from_frequencies(word_freq)

plt.figure(figsize=(12, 6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off') 
plt.title('Plataformas mais escolhidas', fontsize=15)

nome_arquivo_pdf = 'plataformas_wordcloud.pdf'

plt.savefig(nome_arquivo_pdf, bbox_inches='tight')

print(f"\nWordCloud salvo com sucesso em '{nome_arquivo_pdf}' no formato PDF.")

plt.show()
```
